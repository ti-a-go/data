# Create Frequency Table of a VARIABLE

```python
df
```

```python
tag_freq = pd.DataFrame()
tag_freq["tag_"] = df["tag_"].value_counts().index
tag_freq["value_counts"] = df["tag_"].value_counts().values
tag_freq
```

```python
frequency = df["tag_"].value_counts(normalize=True) * 100
tag_freq["frequency"] = frequency.values
tag_freq
```
