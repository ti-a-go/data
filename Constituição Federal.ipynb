# Limpeza de Dados

## Remover Quebra de Linha Dupla

```python
data = data.split("\n\n")
len(data)
```

```python
data[0]
```

## Remover Quebra de Linha

```python
lines = []
for concatened_lines in data:
    lines += concatened_lines.split("\n")
len(lines)
```

```python
lines[0]
```

```python
for line in lines:
    if line == "":
        print("Oops!")
```

```python
lines = lines[1:]
lines
```

## Remover Espaços em Branco no Início e Final de Cada Linha

```python
lines = [line.strip() for line in lines]
lines
```

# Salvar Linhas em um Arquivo JSON

```python
import json
```

```python
content = json.dumps(lines)
content
```

```python
import os

file_name = "Constituição Federeal.json"
with open(os.path.join(nlp.DATA_FOLDER, file_name), "w") as file:
    file.write(content)
    file.close()
```

## Artigos

```python
articles = []
for line in lines:
    if line.startswith("Art. "):
        articles.append(line)
articles
```

## Tarefas

Processar os Artigos usando o modelo de Português do Spacy

```python
doc = nlp.create_doc(articles[0])
doc
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    if token.pos_ == "VERB":
        print(token.text)
```

```python
df = nlp.create_tokens_df(doc)
df.head(10)
```
